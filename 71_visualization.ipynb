# regplot (sns)

```python
tip_df = sns.load_dataset('tips')
tip_df.head()
```

```python
sns.regplot(x=tip_df['total_bill'], y=tip_df['tip'], scatter=False)
```

```python
# ci 속성
sns.regplot(x=tip_df['total_bill'], y=tip_df['tip'], scatter=False, ci=None)
```

# boxplot   
데이터가 많을수록 잘 대표한다.    
이상치 탐지, 통계량 확인, 분포 확인

```python
ev_df = pd.read_csv(
    './data/EV_charge.csv' 
)
ev_df.info()
```

```python
ev_df.head()
```

```python
# object data : x = weekday, platform
# numerical data :  y = kwhTotal, dollars
```

```python
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df)
```

# boxplot + stripplot

```python
# 원본 데이터 + 통계량을 같이 표현한다 >  boxplot + stripplot    
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df)
sns.stripplot(x='weekday', y='kwhTotal', data=ev_df)
```

# swarmplot

boxplot이 만능이 아니다.

swamplot : 값이 겹치지 않는다 -> 분포를 잘 보여준다 . 성능 저하 고려

```python
# 원본 데이터 + 통계량을 같이 표현한다 >  boxplot + stripplot    
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df)
sns.stripplot(x='weekday', y='kwhTotal', data=ev_df)
sns.swarmplot(x='weekday', y='kwhTotal', data=ev_df,color='grey', alpha=0.4)
```

# boxplot 2

```python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df, order=weekday_order)
```

```python
weekday_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sns.boxplot(x='weekday', y='kwhTotal', data=ev_df,
             order=weekday_order, hue='platform')
```

```python
# plotly boxplot
px.box(
    ev_df, x = 'weekday', y='kwhTotal',
    width=500, height=500
)
```

```python
weekday_orders = {'weekday': weekday_order}
```

```python
# plotly boxplot
px.box(
    ev_df, x = 'weekday', y='kwhTotal',
    width=500, height=500,
    category_orders=weekday_orders, color = 'platform'
).show()
```
